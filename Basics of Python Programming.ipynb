## Exercise 5: Check for Palindrome

Write a Python function called `is_palindrome` that takes a string as input and returns `True` if the string is a palindrome, and `False` otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.

### Example:
- Input: `"racecar"`
- Expected Output: `True`


```python
def is_prime(num):
    """Check if a number is prime."""
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage:
number = 29
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
import random

def product_of_random_numbers():
    """Generate two random numbers, ask the user for their product, and check the answer."""
    num1 = random.randint(1, 100)
    num2 = random.randint(1, 100)
    print(f"What is the product of {num1} and {num2}?")
    user_answer = int(input("Your answer: "))
    
    actual_product = num1 * num2
    
    if user_answer == actual_product:
        print("Correct! Well done!")
    else:
        print(f"Incorrect. The correct answer is {actual_product}.")

# Example usage:
product_of_random_numbers()
def squares_of_even_numbers():
    """Print the squares of even numbers between 100 and 200."""
    for num in range(100, 201):
        if num % 2 == 0:
            print(f"The square of {num} is {num ** 2}")

# Example usage:
squares_of_even_numbers()
def word_counter(input_text):
    """Count the number of occurrences of each word in a given text."""
    words = input_text.split()
    word_count = {}
    
    for word in words:
        word = word.strip(",.!?").lower()
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    for word, count in word_count.items():
        print(f"'{word}': {count}")

# Example usage:
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counter(input_text)
def is_palindrome(input_string):
    """Check if a given string is a palindrome."""
    cleaned_string = ''.join(e for e in input_string if e.isalnum()).lower()
    return cleaned_string == cleaned_string[::-1]

# Example usage:
input_string = "A man, a plan, a canal, Panama"
print(is_palindrome(input_string))  # Output: True
```
